```python
import tensorflow as tf
```

```python
# Create features(using tensors)
X = tf.constant([-7.0,-4.0,-1.0,2.0,5.0,8.0,11.0,14.0])

# Create labels(using labels)
y = tf.constant([3.0,6.0,9.0,12.0,15.0,18.0,21.0,24.0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(X,y)
plt.show()
```

```python
print(X.shape)
print(y.shape)
```

```python
# SETTING THE RANDOM STATE
# tf.random.set_seed(42)

#  CREATE THE MODEL USING SEQUENTIAL API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)   # HERE 1 IS NUMBER OF VARIABLES IN THE INPUT LAYER
])

model
```

```python
# COMPILE THE MODEL
model.compile(loss=tf.keras.losses.mse,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mse'])
```

```python
# FIT THE MODEL

# X should be always expected in 2 Dimesions
model.fit(tf.expand_dims(X,axis=-1),y,epochs=20)
```

```python
# SAVING ALL THE MSE VALUES IN A VARIABLE
mse_values = model.history.history['mse']
mse_values
```

```python
# PREDICTING THE VALUES
Y_Pred = model.predict([X])
Y_Pred
```

```python
# SCATTER PLOT BETWEEN ACTUALS AND PREDICTIONS

plt.scatter(X,y)
plt.plot(X,y,color='red')
plt.scatter(X,Y_Pred)
plt.plot(X,Y_Pred,color='green')
plt.ylabel('y and y predictions')
plt.xlabel('epochs')
plt.title('Actuals VS Predictions')
plt.show()
```

```python
# SEE THE ERROR HOW IT IS REDUCED

plt.scatter(range(1,21),mse_values)
plt.plot(range(1,21),mse_values,color='green')
plt.xlabel('epochs')
plt.ylabel('mse')
plt.show()
```

```python
# HERE WE HAVE TO INCREASE THE ITERATIONS SO THAT ERROR SHOULD BE LOW
# SO NOW LETS TRY TO RUN WITH 100 EPOCHS

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mse,
               optimizer=tf.keras.optimizers.SGD(),
               metrics=['mse'])

model.fit(tf.expand_dims(X,axis=-1),y,epochs=100)
```

```python
mse_values = model.history.history['mse']
mse_values
```

```python
Y_Pred1 = model.predict(X)
Y_Pred1
```

```python
plt.scatter(X,y)
plt.plot(X,y,color='red')
plt.scatter(X,Y_Pred1)
plt.plot(X,Y_Pred1,color='green')
plt.plot(X,y,color='red')
plt.show()
```

```python
# NOW PLOT AND OBSERVE THE REDUCING THE ERROR

plt.scatter(range(1,101),mse_values)
plt.plot(range(1,101),mse_values,color='green')
plt.xlabel('epochs')
plt.ylabel('MSE')
plt.show()
```

```python
plt.scatter(X,y)
plt.plot(X,y,color='green')
plt.scatter(X,Y_Pred)
plt.plot(X,Y_Pred,color='red')
plt.scatter(X,Y_Pred1)
plt.plot(X,Y_Pred1,color='blue')
plt.xlabel('epochs')
plt.show()
```

```python
# IF WE ADD SOME MORE EPOCHS THEN BLUE LINE WE GET SOME MORE CLOSER TO THE GREEN LINE
# NOW LETS FIT THE LINE WITH 200 EPOCHS

# HERE WE HAVE TO INCREASE THE ITERATIONS SO THAT ERROR SHOULD BE LOW
# SO NOW LETS TRY TO RUN WITH 100 EPOCHS

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mse,
               optimizer=tf.keras.optimizers.SGD(),
               metrics=['mse'])

model.fit(tf.expand_dims(X,axis=-1),y,epochs=200)
```

```python
mse_values = model.history.history['mse']
mse_values
```

```python
Y_Pred2 = model.predict(X)
Y_Pred2
```

```python
plt.scatter(X,y)
plt.plot(X,y,color='red')
plt.scatter(X,Y_Pred2)
plt.plot(X,Y_Pred2,color='green')
plt.plot(X,y,color='red')
plt.show()
```

```python
plt.scatter(X,y)
plt.plot(X,y,color='green')
plt.scatter(X,Y_Pred)
plt.plot(X,Y_Pred,color='red')
plt.scatter(X,Y_Pred1)
plt.plot(X,Y_Pred1,color='blue')
plt.scatter(X,Y_Pred2)
plt.plot(X,Y_Pred2,color='pink')
plt.xlabel('epochs')
plt.show()
```
